## Polymorphism

### Polymorphism in functions
One function can work on multiple datatypes or classes

```python
a = "Utkarsh"
type(a)
```

```python
len(a)
```

```python
b = [1, 2, 3, 4]
type(b)
```

```python
len(b)
```

```python
c = {"a": 1, "b": 2}
type(c)
```

```python
len(c)
```

```python
d = (2, 3, 4, 5, 6)
type(d)
```

```python
len(d)
```
